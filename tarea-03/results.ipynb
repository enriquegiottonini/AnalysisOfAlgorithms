```python
# read benchmark results from csv files and plot them
# first column is the size of algorithm
# second column is the time of algorithm 1
# third column is the time of algorithm 2
# etc.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# read csv files
def read_csv(path):
    df = pd.read_csv(path, sep=',')
    return df

# plot the results
def plot_results(df, title):
    # get the size of the algorithms
    x = df.iloc[:, 0].values
    # get the time of the algorithms
    y = df.iloc[:, 1:].values
    # plot the results
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel('Size N')
    plt.ylabel('Time (s)')
    plt.legend(df.columns[1:])
    plt.show()
```

```python
sums = read_csv('benchmark_sums.csv')
plot_results(sums, 'Sum of numbers')
```

```python
polynomial = read_csv('benchmark_polynomials.csv')
plot_results(polynomial, 'Polynomial')
```

```python
fibb = read_csv('benchmark_fibonacci.csv')
plot_results(fibb, 'Fibonacci')
```

```python
transpose = read_csv('benchmark_transpose.csv')
plot_results(transpose, 'Transpose')
```

```python
multiplication = read_csv('benchmark_multiplication.csv')
plot_results(multiplication, 'Multiplication')
```

```python
permutations = read_csv('benchmark_permutation.csv')
plot_results(permutations, 'Permutation')
```

```python
search = read_csv('benchmark_search.csv')
plot_results(search, 'Search')
```
